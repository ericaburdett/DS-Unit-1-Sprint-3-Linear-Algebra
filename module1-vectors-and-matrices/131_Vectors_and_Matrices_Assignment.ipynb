## 1.8 Find $||g||$ and then find $||h||$. Which is longer?

\begin{align}
\vec{g} = \begin{bmatrix} 1 \\ 1 \\ 1 \\ 8 \end{bmatrix}
\qquad
\vec{h} = \begin{bmatrix} 3 \\3 \\ 3 \\ 3 \end{bmatrix}
\end{align}

```
# Set up vectors g and h
g = np.array([[1], [1], [1], [8]])
h = np.array([[3], [3], [3], [3]])

# Find norms
norm_g = np.linalg.norm(g)
print('Norm g: ', norm_g)

norm_h = np.linalg.norm(h)
print('Norm h: ', norm_h)

```

```
# The norm of a vector is its length, so g is longer.
```

## 2.2 Find the following products: CD, AE, and BA. What are the dimensions of the resulting matrices? How does that relate to the dimensions of their factor matrices?

```
# Setup the matrices
A = np.array([[1, 2],
             [3, 4],
             [5, 6]])
B = np.array([2, 4, 6])
C = np.array([[9, 6, 3],
             [4, 7, 11]])
D = np.array([[1, 0, 0],
             [0, 1, 0],
             [0, 0, 1]])
E = np.array([[1, 3],
             [5, 7]]) 
```

```
# Find CD
CD = np.matmul(C, D)
print(CD)
```

```
# Find AE
AE = np.matmul(A, E)
print(AE)
```

```
# Find BA
BA = np.matmul(B, A)
print(BA)
```

```
# The dimensions of the final matrix equals the number of rows in the first by 
# the number of columns in second
```

## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\

\end{bmatrix}

```
# Set up H and J
H = np.array([[12, 11],
              [7, 10]])

J = np.array([[0, 1, 2],
              [7, 10, 4],
              [3, 2, 0]])
```

```
# Find |H|
detH = np.linalg.det(H)
print(detH)
```

```
# Find |J|
detJ = np.linalg.det(J)
print(detJ)
```
